# Example of how to write PNG files for YOLOv8 training input

```python
from pathlib import Path

DATA_PATH = Path("data") # TODO: Replace with your data path
```

```python
import hyperspy.api as hs

# Load the microscopy dataset
experiment_file = DATA_PATH / "2023081040-700kx MultiFrams 991ms 600F Counting & FFI mode Falcon 700 kx 1819.emd"
signals = hs.load(str(experiment_file))
```

```python
# Plot the gold nanoparticles
signals.plot()
```

```python
# Setup the output directory
image_path = DATA_PATH / "images"
image_path.mkdir(exist_ok=True)

# Write an image file for each time step in the spatiotemporal data
for i, single_image in enumerate(signals):
   single_image.save(str(image_path / f"image{i}.png"))
```
